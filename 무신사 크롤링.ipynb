```python
import chromedriver_autoinstaller
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from urllib.request import Request, urlopen
import os
import MySQLdb
```

```python
db = MySQLdb.connect('localhost', 'root', '1234', 'test')
cur = db.cursor()
sql = "create table musinsa(m_idx int not null auto_increment primary key,m_category TEXT not null,m_path TEXT)"
cur.execute(sql) 
db.commit()
db.close()
```

```python
driver = webdriver.Chrome()
driver.implicitly_wait(3)

q=1

for i in range(1,5):
    
    folder_name = '무신사 옷'
    os.makedirs(folder_name, exist_ok=True)
    
    url = f'https://www.musinsa.com/categories/item/001001?d_cat_cd=001001&brand=&list_kind=small&sort=pop_category&sub_sort=&page={i}'
    driver.get(url)
    time.sleep(8)
    image_area_xpath = '/html/body/div[2]/div[3]/div[11]/div/div[1]/div[2]/div[3]/ul'
    image_area = driver.find_element(By.XPATH, image_area_xpath)
    image_elements = image_area.find_elements(By.CSS_SELECTOR, 'div.list_img')
    
    image_urls = []

    for image_element in image_elements:
        image_img = image_element.find_element(By.TAG_NAME, 'img')
        image_url = image_img.get_attribute('lazyload lazy')
        if image_img.get_attribute('lazyload lazy') is None:
            image_url = image_img.get_attribute('src')
        print(image_url)
        image_urls.append(image_url)
        
    for i in range(len(image_urls)):
            image_url = image_urls[i]
            image_byte = Request(image_url, headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'})
            file_name = f'clothes{q}.jpg'
            file_path = os.path.join(folder_name, file_name)
            
            db = MySQLdb.connect('localhost', 'root', '1234', 'test')
            cur = db.cursor()
            sql = "insert into musinsa (m_category, m_path) values ( %s, %s)"
            data = (folder_name,file_path)
            cur.execute(sql, data)
            db.commit()
            db.close() 
            
            with open(file_path, 'wb') as f:
                f.write(urlopen(image_byte).read())
            q += 1
```
